```python
from matplotlib import pyplot as plt


def display_palette(palette):
    plt.figure(figsize=(15, 5))
    plt.imshow([[color.to_rgb() for color in palette]])
    print([color.to_hex() for color in palette])
```

```python
from delicacy.svglib.colors.hsv import HSVColor
from delicacy.svglib.colors.palette import PaletteGenerator
from delicacy.svglib.colors.palette import (
    analogous,
    monochromatic,
    shade,
    tint,
    complementary,
    triad,
    square,
    jewel,
    pastel,
    earth,
    neon,
)
```

```python
display_palette(PaletteGenerator(analogous).generate(12))
```

```python
display_palette(PaletteGenerator(monochromatic).generate(12))
```

```python
display_palette(PaletteGenerator(tint).generate(12))
```

```python
display_palette(PaletteGenerator(shade).generate(12))
```

```python
display_palette(PaletteGenerator(complementary).generate())
```

```python
display_palette(PaletteGenerator(triad).generate())
```

```python
display_palette(PaletteGenerator(square).generate())
```

```python
display_palette(PaletteGenerator(jewel).generate(12))
```

```python
display_palette(PaletteGenerator(pastel).generate(12))
```

```python
display_palette(PaletteGenerator(earth).generate(12))
```

```python
display_palette(PaletteGenerator(neon).generate(12))
```
